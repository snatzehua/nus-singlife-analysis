```python
import torch
import torch.nn as nn
import torch.optim as optim
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder

# Define categorical columns
client_categorical_cols = ["cltsex", "marryd", "race_desc_map"]
agent_categorical_cols = ["agent_marital", "agent_product_expertise"]

# Define customer and agent features
numerical_customer_features = ["cltage", "household_size", "economic_status", "family_size"]
numerical_agent_features = ["agent_age", "agent_tenure", "annual_premium_cnvrt", "agent_gender_encoded"]

# Initialize OneHotEncoder
client_ohe = OneHotEncoder(sparse_output=False)
agent_ohe = OneHotEncoder(sparse_output=False)

# Apply transformation for gender encoding
merged_df["agent_gender_encoded"] = merged_df["agent_gender"].apply(transform_gender)

# One-hot encode categorical columns
client_ohe_encoded = client_ohe.fit_transform(merged_df[client_categorical_cols])
agent_ohe_encoded = agent_ohe.fit_transform(merged_df[agent_categorical_cols])

# Convert to DataFrame
client_ohe_df = pd.DataFrame(client_ohe_encoded, columns=client_ohe.get_feature_names_out(client_categorical_cols))
agent_ohe_df = pd.DataFrame(agent_ohe_encoded, columns=agent_ohe.get_feature_names_out(agent_categorical_cols))

# Transform `cust_age_at_purchase_grp` to numerical values
age_mapping = {
    "AG01_lt20": 17, "AG02_20to24": 22, "AG03_25to29": 27, "AG04_30to34": 32,
    "AG05_35to39": 37, "AG06_40to44": 42, "AG07_45to49": 47, "AG08_50to54": 52,
    "AG09_55to59": 57, "AG10_60up": 62
}
merged_df["cltage"] = merged_df["cust_age_at_purchase_grp"].map(age_mapping)

# Concatenate one-hot encoded columns
merged_df = pd.concat([merged_df.drop(columns=client_categorical_cols), client_ohe_df], axis=1)
merged_df = pd.concat([merged_df.drop(columns=agent_categorical_cols), agent_ohe_df], axis=1)

# Fill NaN values with the median for specified columns
columns_to_fix = ["agent_age", "household_size", "economic_status", "family_size", "cltage"]
median_values = {col: merged_df[col].median() for col in columns_to_fix}

print("NaN counts before filling missing values:")
print(merged_df.isna().sum()[merged_df.isna().sum() > 0])

# Fill NaNs
merged_df.fillna(median_values, inplace=True)

# Drop rows where key fields have NaNs
merged_df.dropna(subset=["cltdob", "cltpcode"], inplace=True)

print("NaN counts after filling missing values:")
print(merged_df.isna().sum()[merged_df.isna().sum() > 0])

# Track indices before shuffling for later mapping
merged_df = merged_df.reset_index()  # Preserves the original index
indices = merged_df.index.values  # Store the indices

# Define final categorical features
categorical_customer_features = list(client_ohe_df.columns)
categorical_agent_features = list(agent_ohe_df.columns)

# Extract numerical features
X_num = merged_df[numerical_customer_features]  
y_num = merged_df[numerical_agent_features]  

# Extract categorical features
X_cat = merged_df[categorical_customer_features]  
y_cat = merged_df[categorical_agent_features]  

# Apply StandardScaler **only to numerical columns**
scaler_X = StandardScaler()
scaler_y = StandardScaler()

X_num_scaled = scaler_X.fit_transform(X_num)
y_num_scaled = scaler_y.fit_transform(y_num)

# Convert scaled numerical data back to DataFrame
X_num_scaled = pd.DataFrame(X_num_scaled, columns=numerical_customer_features, index=merged_df.index)
y_num_scaled = pd.DataFrame(y_num_scaled, columns=numerical_agent_features, index=merged_df.index)

# Concatenate scaled numerical features with unchanged categorical features
X = pd.concat([X_num_scaled, X_cat], axis=1)
y = pd.concat([y_num_scaled, y_cat], axis=1)

# Convert to NumPy arrays
X = X.values.astype(np.float32)
y = y.values.astype(np.float32)

print("NaNs in X before tensor conversion:", np.isnan(X).sum())
print("NaNs in y before tensor conversion:", np.isnan(y).sum())

# Convert to PyTorch tensors
X_tensor = torch.tensor(X, dtype=torch.float32)
y_tensor = torch.tensor(y, dtype=torch.float32)

# Perform train-test split while keeping track of indices
X_train, X_test, y_train, y_test, train_indices, test_indices = train_test_split(
    X_tensor, y_tensor, indices, test_size=0.2, random_state=42
)

# Save test indices to later retrieve `agntnum`
X_test_indices = test_indices  # These match `X_test`

print(f"Stored {len(X_test_indices)} indices for mapping predictions back to agntnum.")
```

```python
import torch.nn as nn

class AgentMatchingNN(nn.Module):
    def __init__(self, input_dim, output_dim):
        super(AgentMatchingNN, self).__init__()
        self.fc1 = nn.Linear(input_dim, 256)  # Increased neurons
        self.bn1 = nn.BatchNorm1d(256)  # Batch Normalization
        self.relu1 = nn.ReLU()
        
        self.fc2 = nn.Linear(256, 128)
        self.bn2 = nn.BatchNorm1d(128)
        self.relu2 = nn.ReLU()
        
        self.fc3 = nn.Linear(128, 64)
        self.bn3 = nn.BatchNorm1d(64)
        self.relu3 = nn.ReLU()

        self.fc4 = nn.Linear(64, 32)
        self.bn4 = nn.BatchNorm1d(32)
        self.relu4 = nn.ReLU()

        self.fc5 = nn.Linear(32, output_dim)
        
    def forward(self, x):
        x = self.relu1(self.bn1(self.fc1(x)))
        x = self.relu2(self.bn2(self.fc2(x)))
        x = self.relu3(self.bn3(self.fc3(x)))
        x = self.relu4(self.bn4(self.fc4(x)))
        x = self.fc5(x)  # Ensure the model outputs the correct shape
        return x

# Initialize model
input_dim = X_train.shape[1]  # Number of customer features
output_dim = y_train.shape[1]  # Number of agent features
model = AgentMatchingNN(input_dim, output_dim)
```

```python
import torch.nn as nn
import torch.nn.functional as F

class MixedLoss(nn.Module):
    def __init__(self, alpha=1.0, beta=1.0, delta=1.0):
        super(MixedLoss, self).__init__()
        self.huber = nn.HuberLoss(delta=delta)  # Huber loss for numerical values
        self.ce = nn.CrossEntropyLoss()  # Cross-entropy loss for categorical values
        self.alpha = alpha  # Weight for Huber loss
        self.beta = beta  # Weight for Cross-Entropy loss

    def forward(self, y_pred, y_true):
        num_cols = 3  # Adjust based on number of numerical columns
        y_pred_num, y_pred_cat = y_pred[:, :num_cols], y_pred[:, num_cols:]
        y_true_num, y_true_cat = y_true[:, :num_cols], y_true[:, num_cols:].long()  # Ensure categorical labels are integers
        
        # Compute losses
        huber_loss = self.huber(y_pred_num, y_true_num)
        ce_loss = self.ce(y_pred_cat, y_true_cat.argmax(dim=1))  # Convert one-hot to class index
        
        print(f"Huber Loss: {huber_loss.item():.4f}, CE Loss: {ce_loss.item():.4f}")
        
        return self.alpha * huber_loss + self.beta * ce_loss

```

```python
# Initialize the loss function
criterion = nn.SmoothL1Loss()  # Huber Loss
optimizer = torch.optim.Adam(model.parameters(), lr=0.001, weight_decay=1e-5)  # L2 regularization
scheduler = torch.optim.lr_scheduler.StepLR(optimizer, step_size=20, gamma=1)  # Reduce lr every 20 epochs

# Training loop
num_epochs = 200
for epoch in range(num_epochs):
    model.train()
    optimizer.zero_grad()
    y_pred = model(X_train)
    loss = criterion(y_pred, y_train)  # Compute loss
    loss.backward()
    torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)  # Prevent exploding gradients
    optimizer.step()
    scheduler.step()  # Apply learning rate decay

    if (epoch+1) % 10 == 0:
        print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}")
```

```python
model.eval()  # Set model to evaluation mode
with torch.no_grad():
    y_test_pred = model(X_test)
    test_loss = criterion(y_test_pred, y_test)
    print(f"Test Loss: {test_loss.item():.4f}")
    rmse = torch.sqrt(test_loss)
    print(f"Test RMSE: {rmse.item():.4f}")
```
